```python
import pandas as pd
import numpy as np
import json      # library for working with JSON-formatted text strings
import requests  # library for accessing content from web URLs
import pprint    # library for cleanly printing Python data structures
pp = pprint.PrettyPrinter()

import seaborn as sns
```

```python
response = requests.get("https://educationdata.urban.org/api/v1/schools/crdc/discipline/2017/disability/race/sex?fips=6&page=1")
results = response.text
data = json.loads(results)

col_names = data['results'][0].keys()
count = data['count']
print(count)

#fips 6 : pages 2, 3, 4, 5, 6 

# We wanted to see what different columns they had in the data 
print(col_names)
print(type(col_names))

direct_2011 = pd.DataFrame(data['results'])

direct_2011[['fips']]

#direct_2011.isnull().sum()
#direct_2011 = pd.DataFrame(data['results'])
direct_2011
```

```python
#doing a four loop for pages 2, 3, 4, 5, 6

#this is just 2011

direct11_all = pd.DataFrame() 

for i in range(1, 248): 
    
    api = "https://educationdata.urban.org/api/v1/schools/crdc/discipline/2011/disability/race/sex?fips=6&page=" + str(i)
    
    response = requests.get(api)
    results = response.text 
    data = json.loads(results)
    direct_2011 = pd.DataFrame(data['results'])

    
    # Drop unnecessary columns
    direct_2011 = direct_2011.drop(labels=['transfers_alt_sch_disc', 'disability', 'lep'], axis=1)


    
    # Put all of our page data frames on top of each other 
    direct11_all = pd.concat([direct11_all, direct_2011]).reset_index(drop=True)
    
direct11_all[direct11_all[['students_susp_in_sch', 'students_susp_out_sch_single',
                           'students_susp_out_sch_multiple', 'expulsions_no_ed_serv',
                           'expulsions_with_ed_serv', 'expulsions_zero_tolerance',
                           'students_corporal_punish', 'students_arrested', 
                           'students_referred_law_enforce']] < 0] = np.nan 

direct11_all

direct11_all.to_csv('discipline_2011.csv', index=False)
```
