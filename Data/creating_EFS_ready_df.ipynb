# Converting everything except year and ncessch to means for Fixed Effect Modeling 

```python
ap_model = ap.drop(['latitude', 'longitude', 'school_name'], axis=1)
```

```python
mean_cols = ['enrollment', 'AP_oneormore_relative',
       'teachers_per_student', 'security_gaurds_per_students',
       'avg_teacher_salary', '%Asian', '%Black', '%Hispanic',
       '%Indigenous \nAmerican ', '%Multiracial', '%Pacific \nIslander',
       '%White', '%Female', '%Male', 'enrl_AP', 'enrl_AP_science',
       'enrl_AP_math', 'enrl_AP_other', 'enrl_gifted_talented',
       'students_SAT_ACT', 'students_chronically_absent',
       'students_report_harass_dis', 'students_susp_in_sch',
       'students_susp_out_sch_single', 'students_susp_out_sch_multiple',
       'students_arrested']

# get mean across all rows 
for i in mean_cols: 
    mean_name = i + '_mean'
    ap_model[mean_name] = ap_model.groupby('ncessch')[i].transform(np.mean)
    ap_model[i] = ap_model[i] - ap_model[mean_name]
    ap_model = ap_model.drop(mean_name, axis=1)

ap_model
 
```

```python
ap_model.to_csv('ap_model.csv', index=False)
```
