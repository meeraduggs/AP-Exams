# Get totals 

```python
cropped = ap_rates[['ncessch', 'year', 'sex', 'race', 'enrollment_crdc']]
totals = cropped[(cropped['race'] == 99) & (cropped['sex'] == 99)]
totals = totals.drop(['sex', 'race'], axis=1)
totals 
```

```python
race_totals = cropped[cropped['sex'] == 99]
race_totals = race_totals[race_totals['race'] != 99]

race_totals = race_totals.merge(totals, on = ['ncessch', 'year'], how = 'left')
race_totals['race_enrl_%'] = race_totals['enrollment_crdc_x'] / race_totals['enrollment_crdc_y']

race_totals = race_totals.drop(['sex', 'enrollment_crdc_x', 'enrollment_crdc_y'], axis=1)


# Change the race column values to be more easily interpretable 
race_dict = {1 : '%White', 2 : '%Black', 3: '%Hispanic', 4 : '%Asian', 
             5: '%Indigenous \nAmerican ', 
             6: "%Pacific \nIslander", 
             7 : "%Multiracial", 8 : "%Nonresident alien", 9 : "%Unknown", 
             20 : "%Other", 99 : "%Total"}
race_totals = race_totals.replace({"race": race_dict})

race_totals
```

```python
race_totals[race_totals['race_enrl_%'].isna()]
```

```python
race_pct = race_totals.pivot(index=['ncessch', 'year'], columns='race', values='race_enrl_%').reset_index() 
race_pct
```

```python
race_pct
```

```python
race_pct[race_pct['race_enrl_%'].isna()]
```

```python
race_pct.to_csv('race_percentages.csv', index=False)
```

```python
sex_totals = cropped[cropped['race'] == 99]
sex_totals = sex_totals[sex_totals['sex'] != 99]

sex_totals = sex_totals.merge(totals, on = ['ncessch', 'year'], how = 'left')
sex_totals['sex_enrl_%'] = sex_totals['enrollment_crdc_x'] / sex_totals['enrollment_crdc_y']

sex_totals = sex_totals.drop(['race', 'enrollment_crdc_x', 'enrollment_crdc_y'], axis=1)

# # Change the race column values to be more easily interpretable 
sex_dict = {1 : '%Male', 2 : '%Female', 9 : "%Unknown", 99 : "%Total"}
sex_totals = sex_totals.replace({"sex": sex_dict})

sex_totals
```

```python
sex_pct = sex_totals.pivot(index=['ncessch', 'year'], columns='sex', values='sex_enrl_%').reset_index() 
sex_pct
```

```python
sex_pct.to_csv('sex_percentages.csv', index=False)
```
