```python
import pandas as pd
import numpy as np
import json      # library for working with JSON-formatted text strings
import requests  # library for accessing content from web URLs
import pprint    # library for cleanly printing Python data structures
pp = pprint.PrettyPrinter()
import missingno as msno 
```

## Get AP, IB, and GT Enrollment—By race and sex

```python
apibgt_2011 = pd.DataFrame() 

for i in range(1, 80): 
    api = "https://educationdata.urban.org/api/v1/schools/crdc/ap-ib-enrollment/2011/race/sex/?fips=6&page=" + str(i)
    if (i  % 10 == 0):
        print(api)
    response = requests.get(api)
    results = response.text 
    data = json.loads(results)
    ap_page = pd.DataFrame(data['results'])
    
    
    # Drop unnecessary columns
    ap_page = ap_page.drop(labels=['disability', 'lep'], axis=1)

    
    # Label all negative values in relevant columns as NaN 
 
    ap_page[ap_page[['enrl_IB', 'enrl_AP', 'enrl_AP_science', 'enrl_AP_math', 
                     'enrl_AP_other', 'enrl_AP_language', 'enrl_gifted_talented']] < 0] = np.nan
    
    # Put all of our page data frames on top of each other 
    apibgt_2011 = pd.concat([apibgt_2011, ap_page]).reset_index(drop=True)


apibgt_2011
```

```python
apibgt_2011.to_csv('APIBGT_2011.csv')
```

## Get SAT and ACT participation—By race and sex

```python
SAT_2011 = pd.DataFrame()  ## CHANGE PAGE RANGE!!! 

for i in range(1, 80): 
    api = "https://educationdata.urban.org/api/v1/schools/crdc/sat-act-participation/2011/race/sex/?fips=6&page=" + str(i)
    if (i  % 10 == 0):
        print(api)
    response = requests.get(api)
    results = response.text 
    data = json.loads(results)
    ap_page = pd.DataFrame(data['results'])
    
    
    # Drop unnecessary columns
    ap_page = ap_page.drop(labels=['fips', 'disability', 'lep'], axis=1)

    
    # Label all negative values in relevant columns as NaN 
 
    ap_page[ap_page[['students_SAT_ACT']] < 0] = np.nan
    
    # Put all of our page data frames on top of each other 
    SAT_2011 = pd.concat([SAT_2011, ap_page]).reset_index(drop=True)


SAT_2011
```
