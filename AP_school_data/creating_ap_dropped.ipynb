```python
import pandas as pd 
import numpy as np 

```

```python
ap_cali = pd.read_csv('AP_Exams_Cali.csv')
```

```python
ap_cali.head()
```

```python
# drop unecessary columns 
ap_cali = ap_cali.drop(['crdc_id', 'fips', 'students_AP_pass_none', 'students_AP_pass_oneormore'], axis=1)

# drop rows with letters in id columns 
ap_cali = ap_cali[ap_cali['ncessch'] != "06CC14100001"]


#going to drop the na values in the students_AP_exam_none 
ap_cali = ap_cali.dropna(axis = 0, subset=('students_AP_exam_none',))

# turn id columns to numeric values 
cols = ['ncessch', 'leaid']
ap_cali[cols] = ap_cali[cols].apply(pd.to_numeric, axis=1)


ap_cali 
```

```python
ap_cali.to_csv('ap_dropped.csv', index = False)
```
